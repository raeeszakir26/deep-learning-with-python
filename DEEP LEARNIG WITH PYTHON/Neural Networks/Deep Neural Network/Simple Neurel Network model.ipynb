```python
''' Your tasked is to create a neural network to predict weather a
student will pass or fail an exam based on two features.Hours study
and previous exam score.The dataset should contain info about
hours study and previous exam score for a group of students along
with their exam outcomes(fail or pass)
'''
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python
Hours_study = [2.5, 3, 4, 5, 1.5, 4.3, 2.1, 5.2, 0.5, 1]
previous_exam_score = [60, 65 ,80, 90, 55, 85, 58, 95, 45, 50]
exam_outcome = ['pass', 'pass', 'pass', 'fail', 'fail' 'pass', 'fail', 'pass', 'fail', 'fail', 'fail']
```

```python
label_encoder = LabelEncoder()
encoded_exam_outcomes = label_encoder.fit_transform(exam_outcome)
```

```python
x = np.column_stack((Hours_study, previous_exam_score))
y = encoded_exam_outcomes
```

```python
clf = MLPClassifier(hidden_layer_sizes=(4,), activation='logistic', max_iter=1000, random_state=42)
clf.fit(x,  y)
```

```python
# inputing the example data

new_student_data = np.array([[2.0, 60]])
new_student_data
```

```python
predicted_outcomes = clf.predict(new_student_data)
```

```python
predicted_outcome_decode = label_encoder.inverse_transform(predicted_outcomes)
```

```python
print('predicted exam outcome for new student : {}'.format(predicted_outcome_decode[0]))
```
